```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import requests

url = 'http://data.insideairbnb.com/the-netherlands/north-holland/amsterdam/2019-12-07/data/listings.csv.gz'
# url = 'http://data.insideairbnb.com/belgium/vlg/antwerp/2019-11-26/data/listings.csv.gz'
data = requests.get(url)
```

```python
from os import getcwd 

file_path = '{current_path}/listings.csv.gz'.format(current_path=getcwd())

with open(file_path, 'wb') as f:
    f.write(data.content)
```

```python
df = pd.read_csv(file_path, sep=",") 
```
